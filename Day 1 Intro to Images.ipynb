```python
import numpy as np
import cv2
```

```python
# Load an color image in grayscale
img = cv2.imread('data/Blue.jpg',0)
```

```python
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.namedWindow('image', cv2.WINDOW_NORMAL)
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Difference is that it uses a key command to close
img = cv2.imread('data/Blue.jpg',0)
cv2.imshow('image',img)
k = cv2.waitKey(0)
if k == 27:         # wait for ESC key to exit
    cv2.destroyAllWindows()
elif k == ord('s'): # wait for 's' key to save and exit
    cv2.imwrite('data/Blue.png',img)
    cv2.destroyAllWindows()
```
